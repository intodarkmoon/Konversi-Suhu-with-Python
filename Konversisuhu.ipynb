```python
# Program menghitung konversi suhu 
print("===========================================")
print("Program menghitung Konversi Suhu")
print("=========================================")

# Pilih konversi
print("Pilih Konversi Suhu")
print("1. Konversi Suhu Celcius")
print("2. Konversi Suhu Fahrenheit")
print("3. Konversi Suhu Kelvin")
print("4. Konversi Suhu Reamur")
pilkonversi = int(input("Masukkan no pilihan konversi = "))
print("======================================================")

# Konversi Suhu Celcius
if(pilkonversi==1) :
    print("Pilihan Konversi (Pilih Salah satu)")
    print("1. Konversi dari Celcius ke Fahrenheit")
    print("2. Konversi dari Celcius ke Kelvin")
    print("3. Konversi dari Celcius ke Reamur")
    print("====================================")
    pilih =int(input("Masukkan no pilihan = "))
    print("====================================")

    if(pilih==1):
        print("Konversi dari Celcius ke Fahrenheit \n")
        cel = float(input("Masukkan nilai suhu celcius = "))
        f= (9/5)*cel + 32
        print(f"Hasil dari suhu {cel} celcius = ", f , "fahrenheit")
    elif(pilih==2):
        print("Konversi dari Celcius ke Kelvin \n")
        cel = float(input("Masukkan nilai suhu celcius = "))
        k = cel + 273
        print(f"Hasil dari suhu {cel} celcius = ", k , "kelvin")
    elif(pilih==3):
        print("Konversi dari Celcius ke Reamur \n")
        cel = float(input("Masukkan nilai suhu celcius = "))
        r = (4/5)*cel
        print(f"Hasil dari suhu {cel} celcius = ", r , "reamur")
    else :
        print("No yg dimasukkan tidak ada di dalam pilihan")

# Konversi Suhu Fahrenheit
elif(pilkonversi == 2) :
    print("Pilihan Konversi (Pilih Salah satu)")
    print("1. Konversi dari Fahrenheit ke Celcius")
    print("2. Konversi dari Fahrenheit ke Kelvin")
    print("3. Konversi dari Fahrenheit ke Reamur")
    print("====================================")
    pilih =int(input("Masukkan no pilihan = "))
    print("====================================")

    if(pilih==1):
        print("Konversi dari Fahrenheit ke Celcius \n")
        fah = float(input("Masukkan nilai suhu Fahrenheit = "))
        c= 5/9*fah-32
        print(f"Hasil dari suhu {fah} fahrenheit = ", c , "celcius")
    elif(pilih==2):
        print("Konversi dari Fahrenheit ke Kelvin \n")
        fah = float(input("Masukkan nilai suhu Fahrenheit = "))
        k = 5/9*fah-32+273
        print(f"Hasil dari suhu {fah} fahrenheit = ", k , "kelvin")
    elif(pilih==3):
        print("Konversi dari Fahrenheit ke Reamur \n")
        fah = float(input("Masukkan nilai suhu fahrenheit = "))
        r = 4/9*fah-32
        print(f"Hasil dari suhu {fah} fahrenheit = ", r , "reamur")
    else :
        print("No yg dimasukkan tidak ada di dalam pilihan")
        
# Konversi Suhu Kelvin
elif(pilkonversi== 3):
    print("Pilihan Konversi (Pilih Salah satu)")
    print("1. Konversi dari Kelvin ke Celcius")
    print("2. Konversi dari Kelvin ke Fahrenheit")
    print("3. Konversi dari Kelvin ke Reamur")
    print("====================================")
    pilih =int(input("Masukkan no pilihan = "))
    print("====================================")

    if(pilih==1):
        print("Konversi dari Kelvin ke Celcius \n")
        kel = float(input("Masukkan nilai suhu kelvin = "))
        c = kel - 273
        print(f"Hasil dari suhu {kel} kelvin = ", c , "celcius")
    elif(pilih==2):
        print("Konversi dari Kelvin ke Fahrenheit \n")
        kel = float(input("Masukkan nilai suhu kelvin = "))
        f = 9/5*kel-273+32
        print(f"Hasil dari suhu {kel} kelvin = ", f , "fahrenheit")
    elif(pilih==3):
        print("Konversi dari Kelvin ke Reamur \n")
        kel = float(input("Masukkan nilai suhu kelvin = "))
        r = (4/5)*kel-273
        print(f"Hasil dari suhu {kel} kelvin = ", r , "reamur")
    else :
        print("No yg dimasukkan tidak ada di dalam pilihan")
        
        
# Konversi Suhu Reamur
elif(pilkonversi==4):
    print("Pilihan Konversi (Pilih Salah satu)")
    print("1. Konversi dari Reamur ke Celcius")
    print("2. Konversi dari Reamur ke Fahrenheit")
    print("3. Konversi dari Reamur ke Kelvin")
    print("====================================")
    pilih =int(input("Masukkan no pilihan = "))
    print("====================================")

    if(pilih==1):
        print("Konversi dari Reamur ke Celcius \n")
        re = float(input("Masukkan nilai suhu reamur = "))
        c = 5/4*re
        print(f"Hasil dari suhu {re} reamur = ", c , "celcius")
    elif(pilih==2):
        print("Konversi dari Reamur ke Fahrenheit \n")
        re = float(input("Masukkan nilai suhu reamur = "))
        f = 9/4*re+2
        print(f"Hasil dari suhu {re} reamur = ", f , "fahrenheit")
    elif(pilih==3):
        print("Konversi dari Reamur ke Kelvin \n")
        re = float(input("Masukkan nilai suhu reamur = "))
        k = 5/4*re+273
        print(f"Hasil dari suhu {re} reamur = ", k , "kelvin")
    else :
        print("No yg dimasukkan tidak ada di dalam pilihan")
        
else :
    print("===============================================")
    print("No yg dimasukkan diluar pilihan konversi suhu")
```
